```python
from sklearn.preprocessing import StandardScaler #need to learn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

### Model fitting

##### Random forest Regressor

```python
x = df.drop(columns=['target'],axis=1)
y = df['target'] 
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
error_score = metrics.r2_score(y_train,y_train_pred)
mea         = metrics.mean_absolute_error(y_train,y_train_pred)


print("R2 Score : ",error_score)
print("-----------------------")
print("Mean Absolute Error: ",mea)
```
